```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Load the IPL match analysis dataset into a Pandas DataFrame
df = pd.read_csv("/content/matches.csv")


# Extract the relevant columns for clustering
# Assuming you have columns like 'Runs', 'Wickets', 'Batting Average', 'Bowling Average', etc.
selected_columns = ['win_by_runs', 'win_by_wickets']
data = df[selected_columns]
print(data)

# Perform feature scaling on the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)

# Perform k-means clustering
k = 3  # Number of clusters
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(scaled_data)

# Get the cluster labels and add them to the DataFrame
df['Cluster'] = kmeans.labels_

# Access the cluster centers (centroids)
centroids = kmeans.cluster_centers_

# Print the cluster centers
for i in range(k):
    print("Cluster", i+1, "Center:")
    print(centroids[i])

colors = ['r', 'g', 'b']
markers = ['o', 's', 'D']
for i in range(k):
    cluster_data = df[df['Cluster'] == i]
    plt.scatter(cluster_data['win_by_runs'], cluster_data['win_by_wickets'], c=colors[i], marker=markers[i], label='Cluster ' + str(i + 1))

plt.xlabel('win_by_runs')
plt.ylabel('win_by_wickets')
plt.title('IPL Match Analysis Clustering')
plt.legend()
plt.show()

# Perform any other analysis or visualization on the clustered data as needed.
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Load the IPL match analysis dataset into a Pandas DataFrame
df = pd.read_csv("/content/matches.csv")

# Extract the relevant columns for linear regression
# Assuming you have columns like 'Runs', 'Wickets', 'Batting Average', 'Bowling Average', etc.
selected_columns = ['win_by_runs', 'win_by_wickets']
data = df[selected_columns]

# Split the data into independent (X) and dependent (y) variables
X = data[['win_by_runs']]  # Independent variable(s)
y = data['win_by_wickets']  # Dependent variable

# Perform linear regression
regression_model = LinearRegression()
regression_model.fit(X, y)

# Make predictions using the regression model
y_pred = regression_model.predict(X)

# Plotting the linear regression line and data points
plt.scatter(X, y, color='b', label='Actual Data')
plt.plot(X, y_pred, color='r', label='Regression Line')
plt.xlabel('win_by_runs')
plt.ylabel('win_by_wickets')
plt.title('IPL Match Analysis Linear Regression')
plt.legend()
plt.show()
```
